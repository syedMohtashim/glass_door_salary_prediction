```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle
import seaborn as sns
```

```python
salary_df = pd.read_csv('proj_data/data_for_salary_prediction.csv')
salary_df.drop(['Unnamed: 0'] ,axis = 1, inplace = True)
salary_df.head()
```

```python
salary_df['Hourly Salary'] = salary_df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
# salary_df.loc[salary_df['Hourly Salary'] == 1].head()
salary_df.head()
```

```python
# Remove 'per hour' from the salary estimate
salary_df['Salary Estimate'] = salary_df['Salary Estimate'].apply(lambda x: x.lower().replace('per hour' , ''))
salary_df.head()
```

```python
salary_df['Employer_provided_salary'] = salary_df['Salary Estimate'].apply(lambda x : 1 if 'employer provided salary:' in x.lower() else 0)
# salary_df.loc[salary_df['Employer_provided_salary'] == 1].head()
salary_df.head()
```

```python
# Remove 'Employer Provided Salary' from the salary Estimate
salary_df['Salary Estimate'] = salary_df['Salary Estimate'].apply(lambda x: x.lower().replace('employer provided salary:' , ''))
salary_df.head()
```

```python
salary_postve = salary_df[salary_df['Salary Estimate'] != '-1']
salary_postve.head()
```

```python
salary_postve['Salary Estimate'] = salary_postve['Salary Estimate'].apply(lambda x: x.split('(')[0])
salary_postve.head()
```

```python
salary_postve['Salary Estimate'] = salary_postve['Salary Estimate'].apply(lambda x: x.replace('k' , '').replace('$' , ''))
salary_postve.head()
```

```python
salary_postve['Minimum salary(in thousands of $)'] = salary_postve['Salary Estimate'].apply(lambda x: int(x.split('-')[0]))
salary_postve.head()
```

```python
salary_postve['Maximum salary(in thousands of $)'] = salary_postve['Salary Estimate'].apply(lambda x: int(x.split('-')[1]))
salary_postve.head()
```
