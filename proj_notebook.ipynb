```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle
import seaborn as sns
```

```python
salary_df = pd.read_csv('proj_data/data_for_salary_prediction.csv')
salary_df.drop(['Unnamed: 0'] ,axis = 1, inplace = True)
salary_df.head(50)
```

# First,we have to clean the Salary estimate

```python
salary_negve = salary_df[salary_df['Salary Estimate'] != '-1']
len(salary_negve)
```

There are 742 instances where '-1' is the salary estimate

#### Now, we have to do some spliting

```python
salary = salary_df['Salary Estimate'].apply(lambda x: x.split('(')[0])
salary
```

##### Remove the K and dollar-sign from the salary-estimate 

```python
salary = salary.apply(lambda x: x.replace('K' , '').replace('$' , ''))
salary
```

#### Now, look at the hourly salary

```python
salary_df['Hourly Salary'] = salary_df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
salary_df.loc[salary_df['Hourly Salary'] == 1].head()
```

```python
# Remove 'per hour' from the salary estimate
salary_df['Salary Estimate'] = salary_df['Salary Estimate'].apply(lambda x: x.lower().replace('per hour' , ''))
salary_df.head()
```

##### There is also some Employer provided salaries, make a column for them

```python
salary_df['Employer_provided_salary'] = salary_df['Salary Estimate'].apply(lambda x : 1 if 'employer provided salary:' in x.lower() else 0)
salary_df.loc[salary_df['Employer_provided_salary'] == 1].head()
```

```python
# Remove 'Employer Provided Salary' from the salary Estimate
salary_df['Salary Estimate'] = salary_df['Salary Estimate'].apply(lambda x: x.lower().replace('employer provided salary:' , ''))
salary_df.head(50)
```

#### Minimum  and Maximum Salary

```python
salary_df['Minimum salary'] = salary.apply(lambda x: int(x.split('-')[0]))
salary_df
```
